```python
import sys
sys.path.append('..')

import gdoc_revisions
from oauth2client.service_account import ServiceAccountCredentials
import ipywidgets as widgets
from IPython.display import display, clear_output

CREDENTIAL_FILE = '../settings/doc-revisions-98c56004e106.json'
scope = ['https://www.googleapis.com/auth/drive']
credentials = ServiceAccountCredentials.from_json_keyfile_name(CREDENTIAL_FILE, scope)
FILE_ID = '1aSAA-ZA8bGvJSpgFhVgJu89EDMPMRm5IrcdeV85JmJE'

gdoc = gdoc_revisions.GoogleDoc(FILE_ID, credentials)
```

```python
# notebook requirements

import pandas as pd
import plotly
plotly.offline.init_notebook_mode()
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
# create pandas dataframe of revisions
df_revisions = pd.DataFrame([r.__dict__ for r in gdoc.revisions])
```

```python
df_revisions.head()
```

```python
# document revision index vs time
data = go.Scatter(
    y = df_revisions.apply(lambda r: getattr(r.operation, 'start_index', None),1),
    x = df_revisions.time,
    mode = 'markers',
    name = 'revisions'
)
layout = go.Layout(
    yaxis=dict(autorange='reversed')
)
fig = go.Figure(data=[data], layout=layout)

data = [data]

iplot(fig)
```
